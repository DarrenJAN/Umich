```julia
"""
`x = nnlsgd(A, b ; mu=0, x0=zeros(size(A,2)), nIters::Int=200)`
Performs projected gradient descent to solve the least squares problem:
``\\argmin_{x \\geq 0} 0.5 \\| b − A x \\|_2`` with nonnegativity constraint.
In:
− `A` `m x n` matrix
− `b` vector of length `m`
Option:
− `mu` step size to use, and must satisfy ``0 < mu < 2 / \\sigma_1(A)^2``
to guarantee convergence,
where ``\\sigma_1(A)`` is the first (largest) singular value.
Ch.5 will explain a default value for `mu`
− `x0` is the initial starting vector (of length `n`) to use.
Its default value is all zeros for simplicity.
− `nIters` is the number of iterations to perform (default 200)
Out:
`x` vector of length `n` containing the approximate LS solution
"""
function nnlsgd(A, b ; mu::Real=0, x0=zeros(size(A,2)), nIters::Int=200)
    
    if (mu == 0) 
        mu = 1. / (maximum(sum(abs.(A),dims=1)) * maximum(sum(abs.(A),dims=2)))
    end
    
    x_next = x0
    
    for i in 1:nIters
        z = x_next - mu * (A' * (A * x_next - b ))
        x_next = max.(0, z)
    end 
    
    return x_next
end 
```

```julia
#Question b
using Random: seed!
using Statistics: mean
seed!(0)
m = 100; n = 50; sigma = 2.0
A = randn(m,n)
xtrue = rand(n) # note that xtrue is non−negative
b = A * xtrue + sigma * randn(m)
x0 = A \ b; @show count(x0 .< 0), minimum(x0) # negative values
x0[x0 .<= 0] .= mean(x0[x0 .> 0]) # reasonable initial nonnegative guess

s = svdvals(A)
mu = 1/(s[1,1]^2)
x_100 = nnlsgd(A, b ; mu, x0=x0,  nIters=400)
x_100[1:3]
```

```julia
#Question c
using Optim # you will likely need to add this package
using LinearAlgebra
lower = zeros(n)
upper = fill(Inf, (n,))
inner_optimizer = GradientDescent()
f = x -> 1/2 * norm(A * x - b)^2 # cost function
function grad!(g, x) # its gradient
g[:] = A' * (A * x - b)
end
results = optimize(f, grad!, lower, upper, x0,Fminbox(inner_optimizer), Optim.Options(g_tol=1e-12))
xnnls = results.minimizer
xnnls[5:7]
```

```julia
#Question d
_,s,_=svd(A)
sigma1=0.1/(s[1]^2)
#mu=sigma1
k=LinRange(0,99,100)

xhat = A \b
y1=zeros(100)
for i=1:100
    y1[i]=log10(norm(nnlsgd(A,b,nIters=i,mu=sigma1)-xhat ))
end

plot(xlabel = "k")
plot!(ylabel = "log10(||xk-xhat||)")
plot!(k,y1, label ="NNLS")
plot!(title="u = 0.1/sigma_1(A)")
savefig("1.png")
```

```julia
#Question d --- One plot 
_,s,_=svd(A)
mu1=0.1/(s[1]^2)
mu2=0.5/(s[1]^2)
mu3=1/(s[1]^2)
mu4=1.9/(s[1]^2)
k=LinRange(0,99,100)

xhat = A \b
y1=zeros(100)
y2=zeros(100)
y3=zeros(100)
y4=zeros(100)
for i=1:100
    y1[i]=log10(norm(nnlsgd(A,b,nIters=i,mu=mu1)-xhat ))
    y2[i]=log10(norm(nnlsgd(A,b,nIters=i,mu=mu2)-xhat ))
    y3[i]=log10(norm(nnlsgd(A,b,nIters=i,mu=mu3)-xhat ))
    y4[i]=log10(norm(nnlsgd(A,b,nIters=i,mu=mu4)-xhat ))
end

plot(xlabel = "k")
plot!(ylabel = "log10(||xk-xhat||)")
plot!(k,y1, label ="mu1")
plot!(k,y2, label ="mu2")
plot!(k,y3, label ="mu3")
plot!(k,y4, label ="mu4")
plot!(title="NNSL for four different mu values")
```
